- [5 points] Show that $||x||_{\infty} = \max_i |x_i|$ satisfies the axioms for a vector norm. (Refer [here](https://www.cis.upenn.edu/~cis515/cis515-11-sl4.pdf#page=2))
 
**Solution**

- **Positivity**:

$|x_i| \geq 0 \implies \max_i |x_i| \geq 0 \implies ||x||_{\infty} > 0$

When $||x||_{\infty} = 0 \implies \max_i |x_i| = 0 \implies |x_i| = 0 \forall i\implies x = \mathbf{0}$

When $x = \mathbf{0}, |x_i| = 0 \forall i \implies \max_i |x_i| = 0 \implies ||x||_{\infty} = 0$

- **Scaling**:

$||\lambda x||_{\infty} = \max_i |\lambda x_i| = |\lambda| \max_i |x_i| = |\lambda| ||x||_{\infty}$

- **Triangle Inequality**:

To show $||x + y||_{\infty} \leq ||x||_{\infty} + ||y||_{\infty}$

We have that $x \leq |x| \implies x_i \leq |x_i|$ and $y_i \leq |y_i| \implies |x_i + y_i| \leq |x_i| + |y_i|$

$\implies \max_i |x_i + y_i| \leq \max_i |x_i| + \max_i |y_i| \implies ||x + y||_{\infty} \leq ||x||_{\infty} + ||y||_{\infty}$

- [10 points] From the definition of a [matrix norm](https://en.wikipedia.org/wiki/Matrix_norm#Matrix_norms_induced_by_vector_norms) show that
$||A||_1 = \max_j \sum_i |a_{ij}|.$

**Solution**

Let's prove that for a matrix $A \in\mathbb{R}^{m\times n}$, $||A||_1 = \max_j \sum_i |a_{ij}|$.

First, recall that $||A||_1 = \sup_{x ≠ 0} \frac{||Ax||_1}{||x||_1} = \sup_{||x||_1 = 1} ||Ax||_1$.

For any $x$ with $||x||_1 = 1$:
\begin{align}
||Ax||_1 &= \sum_{i=1}^m |(Ax)_i| \\
&= \sum_{i=1}^m |\sum_{j=1}^n a_{ij}x_j| \\
&\leq \sum_{i=1}^m \sum_{j=1}^n |a_{ij}||x_j| \\
&= \sum_{j=1}^n |x_j| \sum_{i=1}^m |a_{ij}| \\
&\leq \max_{1 \leq j \leq n} \sum_{i=1}^m |a_{ij}| \sum_{j=1}^n |x_j| \\
&= \max_{1 \leq j \leq n} \sum_{i=1}^m |a_{ij}|
\end{align}

This shows $||A||_1 \leq \max_j \sum_i |a_{ij}|$.

For the reverse inequality, let $k$ be the index achieving the maximum: $\max_j \sum_i |a_{ij}| = \sum_i |a_{ik}|$.
Let $e_k$ be the $k$-th standard basis vector. Then:
\begin{align}
||A||_1 &\geq \frac{||Ae_k||_1}{||e_k||_1} = ||Ae_k||_1 \\
&= \sum_{i=1}^m |a_{ik}| \\
&= \max_{1 \leq j \leq n} \sum_{i=1}^m |a_{ij}|
\end{align}

Therefore, $|A||_1 = \max_j \sum_i |a_{ij}|$.

- [5 points] If $||.||_F$ denotes the [Frobenius matrix norm](https://en.wikipedia.org/wiki/Matrix_norm#Frobenius_norm), show that $||QA||_F = ||A||_F$ if $Q$ is an orthogonal matrix.

**Solution**

\begin{align}
QA &= Q \begin{bmatrix}
    \vert & \vert & \cdots & \vert \\
    a_1   & a_2   & \cdots & a_n   \\
    \vert & \vert & \cdots & \vert
\end{bmatrix} \\
&= \begin{bmatrix}
    \vert & \vert & \cdots & \vert \\
    Q a_1   & Q a_2   & \cdots & Q a_n   \\
    \vert & \vert & \cdots & \vert
\end{bmatrix}
\end{align}

We have that the Frobenius matrix norm can be expressed as
\begin{align}
    ||A||_F = \sum_{j = 1}^n ||a_j||_2 = \sum_{j = 1}^n a_j^T a_j
\end{align}

where $a_j$ are columns of the matrix.
\begin{align}
    \implies ||Q A||_F = \sum_{j = 1}^n ||Q a_j||_2 = \sum_{j = 1}^n (Q a_j)^T (Q a_j) = \sum_{j = 1}^n a_j^T Q^T Q a_j = \sum_{j = 1}^n a_j^T a_j = ||A||_F
\end{align}


### Problem 2 [10 Points]

A manufacturing company produces precision bearings for aerospace applications. Quality control data shows that 2% of bearings have a manufacturing defect. The company employs an ultrasonic inspection method that can correctly identifies defective bearings 95% of the time, and can correctly identify non-defective bearings 98% of the time. A specific bearing undergoes ultrasonic testing and tests positive for a defect. What is the probability that the bearing actually has a manufacturing defect? [Hint: Use Bayes rule]

**Solution**

Let:
- $D$ = bearing has defect (2%)
- $T$ = positive test result
- $P(D)$ = 0.02
- $P(T|D)$ = 0.95 (sensitivity)
- $P(T|D^{c})$ = 0.02 (1 - specificity)

\begin{align*}
P(D|T) &= \frac{P(T|D)P(D)}{P(T|D)P(D) + P(T|D^{c})P(D^{c})} \\
&= \frac{(0.95)(0.02)}{(0.95)(0.02) + (0.02)(0.98)} \\
&= \frac{0.019}{0.019 + 0.0196} \\
&= \frac{0.019}{0.0386} \\
&≈ 0.492 \text{ or about 49.2\%}
\end{align*}

**Solution**

\begin{align}
    F(x) = \frac{1}{2} \left[1 + \mathrm{erf} \left(\frac{x - \mu}{\sigma \sqrt{2}}\right)\right]

\begin{align}
    \mathbb{P}(49 < X_i < 51) = F(51) - F(49) = \mathrm{erf}\left(\frac{\sqrt{2}}{8}\right) \approx 0.19742

#### Subproblem 2: [20 Points]

Consider a mixture of two univariate Gaussian distributions:
$$p(x) = \alpha p_1(x) + (1 - \alpha)p_2(x),$$

where $\alpha\in[0,1]$ is a scalar mixture weight and $p_1(x)$, $p_2(x)$ are univariate Gaussians with different parameters $(\mu_1, \sigma_1) \neq (\mu_2, \sigma_2)$.

a) Derive the expectation and variance of $p(x)$, analytically, using their definitions.

b) For $(\mu_1, \sigma_1) = (0, 0.5)$ and $(\mu_2, \sigma_2) = (1, 0.4)$, write a python script that employs the [univariate Gaussian](https://docs.scipy.org/doc/numpy-1.14.0/reference/generated/numpy.random.multivariate_normal.html) function to generate $1000$ samples from $p(x)$ for $\alpha = [0, 0.1, 0.4, 0.6, 0.8, 1]$ and plot the results in different figures. Make a quantitative comment on the form of the results as $\alpha$ changes.

**Solution**

a) By linearity of the expectation, we have that if $X \sim p$
\begin{align}
    \mu = \mathbb{E}[X] = \alpha\mu_1 + (1 - \alpha)\mu_2.
\end{align}

Now, for the variance, we have that
\begin{align}
    \sigma^2 = \mathrm{Var}[X] &= \mathbb{E}[(X - \mu)^2] \\
                               &= \mathbb{E}[X^2] - \mu^2 \\
                               &= (\alpha\mathbb{E}[X_1^2] + (1 - \alpha)\mathbb{E}[X_2^2]) - \mu^2 \\
                               &= (\alpha (\sigma_1^2 + \mu_1^2) + (1 - \alpha)(\sigma_2^2 + \mu_2^2)) - \mu^2.
\end{align}

Note that above, we utilized that $\sigma_i^2 = \mathbb{E}[X_i^2] - \mu_i^2$.

In the Figure below, you'll notice that with increase in $\alpha$, the distribution shifts further towards 0 (shifts from $\mathcal{N}(\mu_2, \sigma_2)$ towards $\mathcal{N}(\mu_1, \sigma_1)$)).

```python
# b)

import numpy as np
import matplotlib.pyplot as pl
from scipy.stats import norm
mu1, mu2 = 0, 1; sigma1, sigma2 = 0.5, 0.4
N_samples = 1000

# We have a mixture model above that we'll carefully need to sample from:
# One method is to generate samples from X1 ~ N(mu1, sigma1) and X2 ~ N(mu2, sigma2)
# And then have the final samples in X be chosen with probability alpha from X1 and (1 - alpha) from X2
X1 = np.random.normal(mu1, sigma1, N_samples)
X2 = np.random.normal(mu2, sigma2, N_samples)
# Sampling probabilities:
p  = np.random.rand(N_samples)

x_to_plot  = np.linspace(-4, 4, 1001)
alpha_list = np.array([0, 0.1, 0.4, 0.6, 0.8, 1])
pl.figure(figsize = (10, 6))
for i in range(6):
    alpha = alpha_list[i]
    X     = np.select([p<=alpha, p>alpha], [X1, X2])
    # Plotting:
    pl.subplot(2, 3, i + 1)
    pl.hist(X, bins = 25, density = True)
    # Plotting the direct PDF:
    pl.plot(x_to_plot,  alpha * norm.pdf(x_to_plot, mu1, sigma1) 
                      + (1 - alpha) * norm.pdf(x_to_plot, mu2, sigma2),
            '--'
           )
    pl.title(r'$\alpha =$' + str(alpha))
```

## Problem 3 [30 Points]

Consider the bivariate Gaussian random variable $\mathbf{X}=(X_1,X_2)$ distributed according to the joint density:

$$p(x_1, x_2) = \mathcal{N}(    
\begin{bmatrix}
      0  \\
      1
\end{bmatrix},
\begin{bmatrix}
      0.4 & -1  \\
      -1 & 5
\end{bmatrix}
).$$

a) Use the properties of Gaussian distribution to derive the mean and the variance of $p(x_1|x_2 = -1)$.

b) Consider another random variable $\mathbf{Y}=(Y_1,Y_2)$  given by the non-linear transformation of $\mathbf{X}$:
$(y_1, y_2) = (\alpha x_1, x_2 /\alpha + \beta(x_1^2 + \alpha^2))$,
with $\alpha = 1.10$ and $\beta = 0.4$. Write a Python script to generate $N = 5000$ realizations of $\mathbf{X}$ and use them to generate $N=5000$ samples of $\mathbf{Y}$. Plot the samples of $\mathbf{Y}$ along with [Gaussian kernel density estimator](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gaussian_kde.html). Also compute and report the the empirical mean of $\mathbf{Y}$.

**Solution**

We have that $p(x_1, x_2 = -1) = \mathcal{N}(\bar{\mu}, \bar{\sigma})$, where $\bar{\mu}$ and $\bar{\sigma}$ are given by
\begin{align}
    \bar{\mu} &= \mu_1 + \sigma_{12} \sigma_{22}^{-1} (-1 - \mu_2) = 0 + -1(1/5)(-1-1)=0.4 \\
    \bar{\sigma} &= \sigma_{11} - \sigma_{12} \sigma_{22}^{-1} \sigma_{21} = 0.4 - (-1)(1/5)(-1) = 0.2
\end{align}

```python
import numpy as np
import matplotlib.pyplot as pl
from scipy.stats import gaussian_kde

import scipy.stats as sp

# Generate samples from a 2-dimensional Gaussian distribution
num_samples = 5000
mu = np.array([0, 1])
Sigma = np.array([[0.4, -1],
                  [-1,   5]])

# Generating Samples of X:
X = np.random.multivariate_normal(mu, Sigma, num_samples)

# Carrying out the transformation to get Y:
alpha = 1.1
beta  = 0.4

y1 = alpha * X[:, 0]
y2 = X[:, 1] / alpha + beta * (X[:, 0]**2 + alpha**2)
Y  = np.vstack([y1, y2])

# Performing KDE:
density = gaussian_kde(Y)

# Calculating the mean:
ymean = np.mean(Y, 1)
print("The empirical mean vector is given by", ymean)
```

```python
# Setting Grid For Plotting:
xmin = Y[0, :].min() - 0.1
xmax = Y[0, :].max() + 0.1
ymin = Y[1, :].min() - 0.1
ymax = Y[1, :].max() + 0.1

# Plotting:
xgrid        = np.linspace(xmin, xmax, 128)
ygrid        = np.linspace(ymin, ymax, 128)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z            = density.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# Plot the result as an image
pl.figure(figsize=(6,4))
pl.imshow(Z.T.reshape(Xgrid.shape),
          origin='lower', aspect='auto',
          extent=[xmin, xmax, ymin, ymax],
          cmap='jet')

cb = pl.colorbar()
cb.set_label("Density")

# Plotting the points of the distribution with some transparency
pl.plot(Y[0, :], Y[1, :], 'ko', alpha = 0.5, markersize = 1)
```
