### Problem 1 [50 Points]

Consider samples $\{x_i\}_{i = 1}^{i = n}$ from a Gaussian distribution with:
- Known variance: $\sigma^2$
- Unknown mean: $\mu$
- Prior distribution on $\mu$: $\mu \sim \mathcal{N}(m, s^2)$
  
(a) Derive the MLE estimate $\mu_{\text{MLE}}$.

(b) Derive the MAP estimate $\mu_{\text{MAP}}$.

(c) Show that as the number of samples $n$ increases, the MAP estimate converges to the maximum likelihood estimate.

(d) Suppose $n$ is small and fixed. What does the MAP estimator converge to if we increase the prior variance $s^2$?

(e) Suppose $n$ is small and fixed. What does the MAP estimator converge to if we decrease the prior variance $s^2$?

(a) Let's start with the likelihood function for our samples $(x_1, x_2, \dots, x_n)$ from a Gaussian distribution $\mathcal{N}(\mu, \sigma^2)$ with known variance $\sigma^2$ and unknown mean $\mu$.

The likelihood function is:

$$\mathcal{L}(\mu) = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x_i - \mu)^2}{2\sigma^2}\right)$$

To find the MLE, I'll maximize this likelihood with respect to $\mu$. It's more convenient to maximize the log-likelihood:

$$\log \mathcal{L}(\mu) = \sum_{i=1}^{n} \log\left[\frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x_i - \mu)^2}{2\sigma^2}\right)\right]$$

$$\log \mathcal{L}(\mu) = \sum_{i=1}^{n} \left[ -\frac{1}{2}\log(2\pi\sigma^2) - \frac{(x_i - \mu)^2}{2\sigma^2} \right]$$

$$\log \mathcal{L}(\mu) = -\frac{n}{2}\log(2\pi\sigma^2) - \frac{1}{2\sigma^2}\sum_{i=1}^{n}(x_i - \mu)^2$$

To find the value of $\mu$ that maximizes this expression, I'll take the derivative with respect to $\mu$ and set it equal to zero:

$$\frac{d}{d\mu}\log \mathcal{L}(\mu) = \frac{d}{d\mu}\left[-\frac{n}{2}\log(2\pi\sigma^2) - \frac{1}{2\sigma^2}\sum_{i=1}^{n}(x_i - \mu)^2\right]$$

The first term is a constant with respect to $\mu$, so its derivative is zero:

$$\frac{d}{d\mu}\log \mathcal{L}(\mu) = -\frac{1}{2\sigma^2}\frac{d}{d\mu}\sum_{i=1}^{n}(x_i - \mu)^2$$

$$\frac{d}{d\mu}\log \mathcal{L}(\mu) = -\frac{1}{2\sigma^2}\sum_{i=1}^{n}\frac{d}{d\mu}(x_i - \mu)^2$$

$$\frac{d}{d\mu}\log \mathcal{L}(\mu) = -\frac{1}{2\sigma^2}\sum_{i=1}^{n}2(x_i - \mu)(-1)$$

$$\frac{d}{d\mu}\log \mathcal{L}(\mu) = \frac{1}{\sigma^2}\sum_{i=1}^{n}(x_i - \mu)$$

Setting this equal to zero:

$$\frac{1}{\sigma^2}\sum_{i=1}^{n}(x_i - \mu) = 0$$

$$\sum_{i=1}^{n}(x_i - \mu) = 0$$

$$\sum_{i=1}^{n}x_i - n\mu = 0$$

$$n\mu = \sum_{i=1}^{n}x_i$$

$$\mu_{MLE} = \frac{1}{n}\sum_{i=1}^{n}x_i = \bar{x}$$

Therefore, the maximum likelihood estimate for μ is the sample mean:

$$\mu_{MLE} = \bar{x}$$

(b) The posterior distribution is proportional to the product of the likelihood and the prior:

$$p(\mu|x_1,...,x_n) \propto L(\mu) \times p(\mu)$$

Where:
- $L(\mu) = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x_i - \mu)^2}{2\sigma^2}\right)$
- $p(\mu) = \frac{1}{\sqrt{2\pi s^2}} \exp\left(-\frac{(\mu - m)^2}{2s^2}\right)$

Taking the log of the posterior:

$$\log p(\mu|x_1,...,x_n) = \log L(\mu) + \log p(\mu) + C$$

$$= -\frac{1}{2\sigma^2}\sum_{i=1}^{n}(x_i - \mu)^2 - \frac{1}{2s^2}(\mu - m)^2 + C'$$

Differentiating with respect to μ and setting to zero:

$$\frac{d}{d\mu}\log p(\mu|x_1,...,x_n) = \frac{1}{\sigma^2}\sum_{i=1}^{n}(x_i - \mu) - \frac{1}{s^2}(\mu - m) = 0$$

Solving for μ:

$$\frac{n}{\sigma^2}\mu = \frac{1}{\sigma^2}\sum_{i=1}^{n}x_i + \frac{1}{s^2}(m - \mu)$$

$$\frac{n}{\sigma^2}\mu + \frac{1}{s^2}\mu = \frac{1}{\sigma^2}\sum_{i=1}^{n}x_i + \frac{m}{s^2}$$

$$\mu\left(\frac{n}{\sigma^2} + \frac{1}{s^2}\right) = \frac{1}{\sigma^2}\sum_{i=1}^{n}x_i + \frac{m}{s^2}$$

Therefore:

$$\mu_{MAP} = \frac{\frac{1}{\sigma^2}\sum_{i=1}^{n}x_i + \frac{m}{s^2}}{\frac{n}{\sigma^2} + \frac{1}{s^2}} = \frac{\frac{n\bar{x}}{\sigma^2} + \frac{m}{s^2}}{\frac{n}{\sigma^2} + \frac{1}{s^2}}$$

This can be rewritten as:

$$\mu_{MAP} = \frac{\frac{s^2}{s^2 + \sigma^2/n} \bar{x} + \frac{\sigma^2/n}{s^2 + \sigma^2/n} m}{\frac{s^2}{s^2 + \sigma^2/n} + \frac{\sigma^2/n}{s^2 + \sigma^2/n}} = \frac{s^2n\bar{x} + \sigma^2m}{s^2n + \sigma^2}$$

(c) Starting with the MAP estimate expression:

$$\mu_{MAP} = \frac{s^2n\bar{x} + \sigma^2m}{s^2n + \sigma^2}$$

As n approaches infinity:

$$\lim_{n \to \infty} \mu_{MAP} = \lim_{n \to \infty} \frac{s^2n\bar{x} + \sigma^2m}{s^2n + \sigma^2}$$

I can factor out the dominant terms with n:

$$\lim_{n \to \infty} \mu_{MAP} = \lim_{n \to \infty} \frac{s^2n\bar{x} + \sigma^2m}{s^2n + \sigma^2} = \lim_{n \to \infty} \frac{s^2n\bar{x}(1 + \frac{\sigma^2m}{s^2n\bar{x}})}{s^2n(1 + \frac{\sigma^2}{s^2n})}$$

As n approaches infinity, the terms $\frac{\sigma^2m}{s^2n\bar{x}}$ and $\frac{\sigma^2}{s^2n}$ approach zero:

$$\lim_{n \to \infty} \mu_{MAP} = \lim_{n \to \infty} \frac{s^2n\bar{x}(1 + 0)}{s^2n(1 + 0)} = \lim_{n \to \infty} \frac{s^2n\bar{x}}{s^2n} = \lim_{n \to \infty} \bar{x} = \bar{x}$$

Since $\mu_{MLE} = \bar{x}$, we have:

$$\lim_{n \to \infty} \mu_{MAP} = \mu_{MLE}$$

This demonstrates that as the number of samples increases, the influence of the prior diminishes, and the MAP estimate converges to the maximum likelihood estimate.

(d) Starting with the MAP estimate:

$$\mu_{MAP} = \frac{s^2n\bar{x} + \sigma^2m}{s^2n + \sigma^2}$$

As s² approaches infinity:

$$\lim_{s^2 \to \infty} \mu_{MAP} = \lim_{s^2 \to \infty} \frac{s^2n\bar{x} + \sigma^2m}{s^2n + \sigma^2}$$

I can factor out s² from both numerator and denominator:

$$\lim_{s^2 \to \infty} \mu_{MAP} = \lim_{s^2 \to \infty} \frac{s^2(n\bar{x}) + \sigma^2m}{s^2(n) + \sigma^2} = \lim_{s^2 \to \infty} \frac{n\bar{x} + \frac{\sigma^2m}{s^2}}{n + \frac{\sigma^2}{s^2}}$$

As s² approaches infinity, the terms $\frac{\sigma^2m}{s^2}$ and $\frac{\sigma^2}{s^2}$ approach zero:

$$\lim_{s^2 \to \infty} \mu_{MAP} = \frac{n\bar{x} + 0}{n + 0} = \frac{n\bar{x}}{n} = \bar{x}$$

Since $\bar{x} = \mu_{MLE}$, we have:

$$\lim_{s^2 \to \infty} \mu_{MAP} = \mu_{MLE}$$

Therefore, as the prior variance s² increases (making the prior more diffuse or less informative), the MAP estimator converges to the maximum likelihood estimator, even when n is small and fixed. This happens because a larger prior variance means we're placing less confidence in our prior belief about μ, causing the likelihood from the observed data to dominate the posterior.

(e) Starting with the MAP estimate:

$$\mu_{MAP} = \frac{s^2n\bar{x} + \sigma^2m}{s^2n + \sigma^2}$$

As s² approaches zero:

$$\lim_{s^2 \to 0} \mu_{MAP} = \lim_{s^2 \to 0} \frac{s^2n\bar{x} + \sigma^2m}{s^2n + \sigma^2}$$

When s² is very small, the terms containing s² become negligible:

$$\lim_{s^2 \to 0} \mu_{MAP} = \lim_{s^2 \to 0} \frac{s^2n\bar{x} + \sigma^2m}{s^2n + \sigma^2} = \frac{0 + \sigma^2m}{\sigma^2} = m$$

Therefore:

$$\lim_{s^2 \to 0} \mu_{MAP} = m$$

As the prior variance s² approaches zero, the MAP estimator converges to the prior mean m. This makes intuitive sense because a very small prior variance indicates high confidence in our prior belief about μ. When we're extremely certain about our prior (s² → 0), we essentially ignore the observed data, regardless of the sample size, and our posterior estimate becomes identical to our prior mean.

### Problem 2 [50 Points]

Consider the Bayesian linear regression model with
$p(y\mid x,w) = \mathcal{N}(y\mid w^{T}\phi(x),\alpha^{-1}I), \ p(w)=\mathcal{N}(w\mid 0,\beta^{-1}I),$

with $\alpha = 0.5, \beta  = 0.1$. Generate a set of $N = 400$ noisy observations by uniformly sampling

$$y(x) = \sin(\pi x) + \sin(2 \pi x) + \sin(5 \pi x)\quad x\in[-1,1]$$

Once you've created the observations, perturb the data with a normal distribution with standard deviation set to 10% of the data. Compute the MLE and MAP estimates for the weights $w$$using different types and numbers of features $\phi(x)$:
- Monomial basis: $\phi(x) = \{1, x, x^2, x^3, \dots, x^M\}$
- Fourier basis: $\phi(x) = \{0, 1, \sin(\pi x), \cos(\pi x), \sin(2\pi x), \cos(2\pi x), \dots, \sin(M\pi x), \cos(M\pi x)\}$ 
- [Legendre](https://en.wikipedia.org/wiki/Legendre_polynomials) basis: $\{P_0(x), P_1(x), P_2(x), P_3(x), \dots, P_M(x)\}$, where $P_0(x) = 1, P_1(x) = x$, and subsequent polynomials can be generated by the recursion $(n + 1) P_{n+1} (x) = (2n + 1) x P_n(x) - n P_{n - 1}(x).$

For the case $M=5$, plot the data, the mean predictions corresponding to the MLE and MAP estimates for $w$, and $100$ samples from the predictive posterior distribution.

Which set of features works best for this function and why? 

### Solution:
We are given the likelihoods and priors
\begin{align}
    p(y\lvert x,w) = \mathcal{N}(y\mid w^{T}\phi(x),\alpha^{-1}I), \ p(w)=\mathcal{N}(w\mid 0,\beta^{-1}I), \ \alpha=0.5, \beta = 0.1
\end{align}

We will construct a basis functions matrix as an $(N \times m)$ matrix $\Phi$, where m denotes the chosen order for the basis.
\begin{align}
    \Phi =
    \begin{bmatrix}
    \phi_1(x_1) & \phi_2(x_1) & \phi_3(x_1) & \dots  & \phi_m(x_1) \\
    \phi_1(x_2) & \phi_2(x_2) & \phi_3(x_2) & \dots  & \phi_m(x_2) \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    \phi_1(x_N) & \phi_2(x_N) & \phi_3(x_N) & \dots  & \phi_m(x_N) \\
    \end{bmatrix}
\end{align}

Now, we have that the MLE estimate satisfies

\begin{align}
w_{MLE} &= \mathrm{argmax}_{w} p(y\lvert x,w) = \mathrm{argmax}_{w} \mathcal{N}(y\mid w^{T}\phi(x),\alpha^{-1}I) \\
&\propto \mathrm{argmax}_{w} \exp(-(y - \Phi w)^T \alpha (y - \Phi w))
\end{align}

This can be determined by minimizing $(y - \Phi w)^T (y - \Phi w)$ by setting the derivative to zero
\begin{align}
    \frac{\partial}{\partial w} ((y - \Phi w)^T (y - \Phi w)) &= \frac{\partial}{\partial w} (y^T y - 2 w^T \Phi^T y + w^T \Phi^T \Phi w) \\
    &= -2 \Phi^T y + 2\Phi^T \Phi w = 0 \\
\implies w_{MLE} &= (\Phi^T \Phi)^{-1} \Phi^T y
\end{align}

The MAP estimate is given by

\begin{align}
    w_{MAP} = \mathrm{argmax} p(w | x, y) \propto p(y|x, w) p(w)
                                         &\propto \exp\left(-\alpha \frac{\alpha}{2} (y - \Phi w)^T (y - \Phi w) - \frac{\beta}{2} w^T w \right)  
\end{align}

Expanding the term inside the exponential, we can show that the posterior is a Gaussian as well by "completing the square".

\begin{align}
    \frac{\alpha}{2} (y - \Phi w)^T (y - \Phi w) + \frac{\beta}{2} w^T w = \alpha y^T y - 2 \alpha w^T \Phi^T y + w^T (\alpha \Phi^T \Phi + \beta I) w = (w - \mu)^T \Lambda (w - \mu) 
\end{align}

By determining $\mu$ and $\Lambda$ by matching terms, we can have the mean and precision of the underlying Gaussian distribution as

\begin{align}
    \Lambda &= \alpha \Phi^T \Phi + \beta I \\
    \mu     &= \alpha \Lambda^{-1} \Phi^T y
\end{align}

Using the least squares solution now, we have that
\begin{align}
    w_{MAP} = \left(\Phi^T \Phi + \frac{\beta}{\alpha} I \right)^{-1} \Phi^T y
\end{align}

```python
import jax.numpy as jnp
from jax import random
import matplotlib.pylab as plt
%matplotlib inline

# Number of datapoints:
N = 400
# Generating points in x:
rng_key = random.PRNGKey(0)
k1, k2  = random.split(rng_key)
x       = 2 * random.uniform(k1, (N,)) - 1
# Defining the datasets:
f = lambda x:jnp.sin(jnp.pi * x) + jnp.sin(2 * jnp.pi * x) + jnp.sin(5 * jnp.pi * x)
y = f(x)
# Adding noise:
y = y + 0.1 * y.std() * random.normal(k2, (N,))
```

```python
# Visualizing the data:
x_plot = jnp.linspace(-1, 1, 401)
plt.figure(dpi = 150)
plt.plot(x_plot, f(x_plot), 'k--', label = 'Underlying Analytic Distribution')
plt.plot(x, y, 'o', label = 'Data Points')
plt.legend(fontsize = 8)
plt.title(r'$y_1(x) = \sin(\pi x) + \sin(2 \pi x) + \sin(5 \pi x)$')
```

```python
class BayesianLinearRegression(object):
    def __init__(self, X, y, basis = 'monomial', 
                 basis_order = 2, alpha = 1.0, beta = 1.0):
        
        self.X = X.reshape(-1, 1)
        self.y = y.reshape(-1, 1)
        # Getting the feature matrix:
        self.basis       = basis
        self.basis_order = basis_order
        self.Phi         = self.return_feature_matrix(self.X)
        # Store precisions
        self.alpha = alpha
        self.beta  = beta
        # Small jitter for numerically stable matrix inversion
        self.jitter = 1e-8
        
    
    def return_feature_matrix(self, X):
        basis       = self.basis
        basis_order = self.basis_order
        # Creating the basis function matrix:
        if(basis == 'monomial'):
            Phi = (X.reshape(-1, 1))**(jnp.arange(basis_order + 1).reshape(1, -1))

        elif(basis == 'fourier'):
            Phi = jnp.hstack([jnp.zeros([X.size, 1]), 
                              jnp.ones([X.size, 1])])
            # Notion of basis_order in this maybe a little strange since we are
            # considering sin and cos so it's technically 2 * basis_order (??)
            for i in range(1, basis_order + 1):
                Phi = jnp.hstack([Phi, 
                                  jnp.sin(i * jnp.pi * X),
                                  jnp.cos(i * jnp.pi * X)
                                 ])
                
        elif(basis == 'legendre'):
            Phi = jnp.hstack([jnp.ones([X.size, 1]), X])
            # Now using the recurrence relation
            # (n + 1) P_{n + 1}(x) = (2n + 1) x P_n(x) - n P_{n-1}(x)
            for i in range(2, basis_order + 1):
                P_i_minus_one = Phi[:, -1].reshape(-1, 1)
                P_i_minus_two = Phi[:, -2].reshape(-1, 1)
                P_i           = (1 / i) * (  (2 * i - 1) * X * P_i_minus_one
                                           - (i - 1) * P_i_minus_two
                                          )
                # Updating the feature matrix:
                Phi = jnp.hstack([Phi, P_i])
        
        return Phi
        
    # Computes the MLE estimate of the regression weights
    def fit_MLE(self):
        PhiTPhi_inv = jnp.linalg.inv(  jnp.matmul(self.Phi.T, self.Phi)
                                     + self.jitter * jnp.eye(self.Phi.shape[1])
                                    )
        PhiTy       = jnp.matmul(self.Phi.T, self.y)
        w_MLE       = jnp.matmul(PhiTPhi_inv, PhiTy)
        self.w_MLE  = w_MLE
        
        return w_MLE

    # Computes the MAP estimate of the regression weights
    def fit_MAP(self):
        Lambda     = jnp.matmul(self.Phi.T, self.Phi) + \
                     (self.beta/self.alpha) * jnp.eye(self.Phi.shape[1])
        Lambda_inv = jnp.linalg.inv(Lambda)
        PhiTy      = jnp.matmul(self.Phi.T, self.y)
        mu         = jnp.matmul(Lambda_inv, PhiTy)
        self.w_MAP = mu
        # Storing parameters for later use:
        self.Lambda_inv = Lambda_inv
        
        return mu, Lambda_inv

    # Computes the predictive posterior distribution
    def predictive_distribution(self, X_star):
        Phi_star  = self.return_feature_matrix(X_star)
        mean_star = jnp.matmul(Phi_star, self.w_MAP)
        var_star  = jnp.eye(Phi_star.shape[0])/self.alpha + \
                    jnp.matmul(Phi_star, jnp.matmul(self.Lambda_inv, Phi_star.T))

        return mean_star, var_star
```

```python
alpha = 0.5
beta  = 0.1
# Order of Basis:
basis_order = 5
# Defining the models:
model1 = BayesianLinearRegression(x, y, 'monomial', basis_order, alpha, beta)
model2 = BayesianLinearRegression(x, y, 'fourier', basis_order, alpha, beta)
model3 = BayesianLinearRegression(x, y, 'legendre', basis_order, alpha, beta)
```

```python
# Finding the MLE and MAP estimates:
w1_MLE              = model1.fit_MLE()
w1_MAP, Lambda1_inv = model1.fit_MAP()

w2_MLE              = model2.fit_MLE()
w2_MAP, Lambda2_inv = model2.fit_MAP()

w3_MLE              = model3.fit_MLE()
w3_MAP, Lambda3_inv = model3.fit_MAP()
```

```python
# Predict at a set of test points
X_star      = jnp.linspace(-1, 1, 1001)[:,None]

y1_pred_MLE = jnp.matmul(model1.return_feature_matrix(X_star), w1_MLE)
y1_pred_MAP = jnp.matmul(model1.return_feature_matrix(X_star), w1_MAP)

y2_pred_MLE = jnp.matmul(model2.return_feature_matrix(X_star), w2_MLE)
y2_pred_MAP = jnp.matmul(model2.return_feature_matrix(X_star), w2_MAP)

y3_pred_MLE = jnp.matmul(model3.return_feature_matrix(X_star), w3_MLE)
y3_pred_MAP = jnp.matmul(model3.return_feature_matrix(X_star), w3_MAP)
```

```python
# Draw samples from the predictive posterior
num_samples = 100

mean1_star, var1_star = model1.predictive_distribution(X_star)
mean2_star, var2_star = model2.predictive_distribution(X_star)
mean3_star, var3_star = model3.predictive_distribution(X_star)

samples1 = random.multivariate_normal(rng_key, mean1_star.flatten(), 
                                          var1_star, (num_samples,))
samples2 = random.multivariate_normal(rng_key, mean2_star.flatten(), 
                                          var2_star, (num_samples,))
samples3 = random.multivariate_normal(rng_key, mean3_star.flatten(), 
                                          var3_star, (num_samples,))

percentiles1 = jnp.percentile(samples1, jnp.array([5.0, 95.0]), axis=0)
percentiles2 = jnp.percentile(samples2, jnp.array([5.0, 95.0]), axis=0)
percentiles3 = jnp.percentile(samples3, jnp.array([5.0, 95.0]), axis=0)
```

```python
plt.figure(dpi = 150)

plt.subplot(3, 1, 1)
plt.plot(X_star, y1_pred_MLE, label = 'MLE')
plt.plot(X_star, y1_pred_MAP, '--', label = 'MAP')
plt.plot(x, y, 'o', markersize = 2, label = 'Data')
plt.title('$y_1(x)$ (Monomial Basis)', fontsize = 8)
plt.xticks([])
plt.legend(bbox_to_anchor = (-0.1, -0.2))
plt.fill_between(X_star.flatten(), percentiles1[0, :], percentiles1[1, :], color='lightblue')

plt.subplot(3, 1, 2)
plt.plot(X_star, y2_pred_MLE)
plt.plot(X_star, y2_pred_MAP, '--')
plt.plot(x, y, 'o', markersize = 2)
plt.title('$y_1(x)$ (Fourier Basis)', fontsize = 8)
plt.xticks([])
plt.fill_between(X_star.flatten(), percentiles2[0, :], percentiles2[1, :], color='lightblue')

plt.subplot(3, 1, 3)
plt.plot(X_star, y3_pred_MLE)
plt.plot(X_star, y3_pred_MAP, '--')
plt.plot(x, y, 'o', markersize = 2)
plt.title('$y_1(x)$ (Legendre Basis)', fontsize = 8)
plt.fill_between(X_star.flatten(), percentiles3[0, :], percentiles3[1, :], color='lightblue')
```

```python
# Errors:
print('Monomial (MLE) Error:', jnp.linalg.norm(f(X_star) - y1_pred_MLE) / jnp.linalg.norm(f(X_star)))
print('Monomial (MAP) Error:', jnp.linalg.norm(f(X_star) - y1_pred_MAP) / jnp.linalg.norm(f(X_star)))
print('Fourier (MLE) Error:', jnp.linalg.norm(f(X_star) - y2_pred_MLE) / jnp.linalg.norm(f(X_star)))
print('Fourier (MAP) Error:', jnp.linalg.norm(f(X_star) - y2_pred_MAP) / jnp.linalg.norm(f(X_star)))
print('Legendre (MLE) Error:', jnp.linalg.norm(f(X_star) - y3_pred_MLE) / jnp.linalg.norm(f(X_star)))
print('Legendre (MAP) Error:', jnp.linalg.norm(f(X_star) - y3_pred_MAP) / jnp.linalg.norm(f(X_star)))
```

It is noted that $y$ is best approximated by the Fourier basis functions since the basis functions capture the underlying data the best. Polynomial features aren't able to represent the underlying data well enough.
